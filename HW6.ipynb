```
import numpy as np
import matplotlib.pyplot as plt
import math



y=[] # to store the values of the function
x=np.arange(-2,2,0.05) # stroing the x values in the x - numpy array
for i in range(len(x)):
     x[i]=round(x[i],2)

print(x) # printing the x values obtained in the range [-2,2] with step of 0.05

for i in x:
    y.append(round(math.e -i**2 + 0.01*(math.cos(200*i)),4)) # calculating the y value of the function f(x)
    
print(y) # printing the y value 

print('\nMaximum value of the function is :',end=" ") 
print(max(y)) # maximum value of the function

plt.xlabel("x ---->")
plt.ylabel('Y=f(x)')

plt.scatter(x,y)
plt.plot(x,y,'r')
plt.show()
```
